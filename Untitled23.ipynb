1  . CREATE TABLE SalesData (
     CustomerID INT,
     Name VARCHAR(255),
     Age INT,
     City VARCHAR(255),
     PurchaseAmount DECIMAL(10,2),
     PurchaseDate DATE
     );

     INSERT INTO SalesData (CustomerID, Name, Age, City, PurchaseAmount, PurchaseDate)
     ROW_NUMBER() OVER (ORDER BY NEWID()) AS CustomerID,
     CONCAT('Customer_', ROW_NUMBER() OVER (ORDER BY NEWID())) AS Name,
     ABS(CHECKSUM(NEWID()) % 80) + 18 AS Age,
     CASE ABS(CHECKSUM(NEWID()) % 5)
        WHEN 0 THEN 'New York'
        WHEN 1 THEN 'Los Angeles'
        WHEN 2 THEN 'Chicago'
        WHEN 3 THEN 'Houston'
        ELSE 'Miami' END AS City,
     ROUND(RAND(CHECKSUM(NEWID())) * 1000, 2) AS PurchaseAmount,
     DATEADD(day, ABS(CHECKSUM(NEWID()) % 3650), GETDATE()) AS PurchaseDate
     FROM sys.objects a, sys.objects b;

2  . SELECT City, SUM(PurchaseAmount) AS TotalSales
     FROM SalesData
     GROUP BY City;
   . SELECT TOP 5 City, SUM(PurchaseAmount) AS Revenue
     FROM SalesData
     GROUP BY City
     ORDER BY Revenue DESC;

3  . WITH AvgPurchase AS (
     SELECT AVG(PurchaseAmount) AS AvgAmount
     FROM SalesData
     )
     SELECT *
     FROM SalesData
     WHERE PurchaseAmount > (SELECT AvgAmount FROM AvgPurchase);
